```python
import pandas as pd

df2 = pd.read_excel("convBenchmark2excel.xlsx")
df3 = pd.read_excel("convBenchmark3excel.xlsx")
df4 = pd.read_excel("convBenchmark4excel.xlsx")
df5 = pd.read_excel("convBenchmark5excel.xlsx")
```

```python
df2
```

```python
a = pd.concat([df2, df3, df4, df5], axis=0, ignore_index=True)
a.rename(columns={ " time(ms)":"time", ' kernel_size':"kernel_size", ' blocks':"blocks", ' threadsPerBlock':"threadsPerBlock", ' mode':"mode", }, inplace=True)
a['mode'] = a['mode'].replace({' CPU': 1, ' GPU': 2})
```

```python
a.to_excel('combined_data.xlsx', index=False)

```
